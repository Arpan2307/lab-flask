```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel(r"Sample.xlsx",skiprows=2,usecols='B:S',index_col=0)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
df=df.drop(['Adress of Image','Extracted Scale Text','Text in meters'],axis=1)
```

```python
df
```

```python
df.info()
```

```python
X=df.drop(['Actual Distance(m)','Time(min)','Circle x','Circle y','Red icon x','Red icon y'],axis=1)
```

```python
X
```

```python
y=df['Actual Distance(m)']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.corr()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaler1.pkl','wb'))
```

```python
sns.boxplot(X_train)
```

```python
sns.boxplot(X_train_scaled)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
linear=LinearRegression()
linear.fit(X_train_scaled,y_train)
y_pred=linear.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae,score)
print(y_pred)
```

```python
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae,score)
```

```python
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae,score)
```

```python
en=ElasticNet()
en.fit(X_train_scaled,y_train)
y_pred=en.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae,score)
```
